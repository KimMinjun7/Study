```python
cond1= (df1["물품대분류"]=="기저귀")
df2 = df1.loc[cond1].iloc[1:]
df2.head()
```

```python
# 이런식으로 변환을 해줘야한다.
df_time2 = pd.pivot_table(data=df2, index="구매일", values = "구매금액", aggfunc="sum").reset_index()
df_time2.set_index("구매일")
```

```python
# 시계열데이터 분석할때 기본적인 데이터 셋
df_time = pd.pivot_table(data=df2, index="구매일", values="구매금액", aggfunc="sum")
df_time
```

```python
# 1일단위로 묶어서 평균금액 계산
y=df_time["구매금액"].resample("1D").mean()
y
```

```python
# 결측값이 있는지 확인
y.isnull().sum()
```

```python
# 전날 데이터로 결측값을 채워줌
# y.fillna(method="bfill")
# 바로 앞날 데이터로 
y1 = y.fillna(method="ffill")
```

```python
y1
```

```python
# 시계열 분해
    - trend
    - Seasonality
    - Cycle/Residual
```

```python
# 그래프의 크기 조절
from pylab import rcParams

rcParams["figure.figsize"]=15,10
```

```python
# 시계열 모델 생성 (최소샘플이 50개 이상이어야함)
model_series = tsa.seasonal_decompose(y1, model="additive")
# 모델 시각화
fig = model_series.plot()
plt.show()
```

```python
# 모델 구축
train_data, test_data = train_test_split(df_time, test_size=0.2, shuffle=False)
```

```python
test_data.tail()
```

```python
# 정상성 만족하는지 확인
fig, ax = plt.subplots(1,2,figsize=(10,5))
fig.suptitle("Raw Data")
sm.graphics.tsa.plot_acf(train_data.values.squeeze(), lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(train_data.values.squeeze(), lags=30, ax=ax[1])
```

```python
# 1차 차분
diff_train_data = train_data.copy()
diff_train_data = diff_train_data["구매금액"].diff(1)
diff_train_data= diff_train_data.dropna()
print("####### Raw Data ########")
print(train_data)
print("### Differenced Data ###")
print(diff_train_data)
```

```python
# 차분 후 시각화
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.plot(train_data["구매금액"])
plt.legend(["Raw Data (Nonstationary)"])
plt.subplot(212)
plt.plot(diff_train_data,"orange")
plt.legend(["Raw Data (stationary)"])
plt.show()
# 정상성이 만족하는 것을 볼 수 있다.
```

```python
# 정상성 만족하는지 확인
fig, ax = plt.subplots(1,2,figsize=(10,5))
fig.suptitle("Difference Data")
sm.graphics.tsa.plot_acf(diff_train_data.values.squeeze(), lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(diff_train_data.values.squeeze(), lags=30, ax=ax[1])
```

```python
# p,d,q 구하기

auto_arima_model = auto_arima(train_data, start_p=0, start_q=0, max_p=3, max_q=3,
                             m=12, seasonal=True, max_P=3, max_Q=3, trace=True, 
                             D=1,d=1, error_action="ignore", suppress_warnings=True,
                              stepwise=False)
```

```python
auto_arima_model.summary()
```

```python
prediction = auto_arima_model.predict(len(test_data), return_conf_int=True)
predicted_value = prediction[0]
predicted_ub = prediction[1][:0]
predicted_lb = prediction[1][:1]
predict_index = list(test_data.index)
```

```python
def plot_train_test_pred(train,test,pred_df):
    fig,ax = plt.subplots(figsize=(12,7))
    kws = dict(marker='o')
    
    ax.plot(train,label='Train',**kws)
    ax.plot(test,label='Test',**kws)
    ax.plot(pred_df,label='Prediction',ls='--',linewidth=3)

    ax.fill_between(x=pred_df.index,y1=pred_df['lower'],y2=pred_df['upper'],alpha=0.3)
    ax.set_title('Model Validation', fontsize=22)
    ax.legend(loc='upper left')
    fig.tight_layout()
    return fig,ax
```
